```python
import numpy as np 

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.preprocessing import LabelBinarizer
from sklearn.preprocessing import MinMaxScaler
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import random
from numpy import *  
from sklearn.metrics import accuracy_score 
from sklearn.preprocessing import Normalizer
from sklearn.model_selection import train_test_split
from statsmodels.formula.api import ols


from scipy.stats import pearsonr

def get_data(normalise, i): 

   # data_in = genfromtxt('data/abalone.data', delimiter=",") # in case of csv data
    data_in = pd.read_csv ('data/abalone.data')
    print(data_in)
    
    # adding column name to the respective columns
    print('add column names')
    
    data_in.columns =['Sex', 'Length', 'Diameter', 'Height','Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight','Rings']
    print(data_in)
     #one hot coding
    # handle categorical variable
    data_in=pd.get_dummies(data_in,drop_first=False)
    print('sex'))
    print(data_in
    # dropping extra column
    #data_in= data_in.drop('Sex',axis=1)
    # concatation of independent variables and new cateorical variable.
    #data_in=pd.concat([data_in,sex],axis=1) 
    
    #Create Correlation Matrix

    corr_mat = data_in.corr(method='pearson')
    print("corr matrix")
    print(corr_mat)
    
    #sort the corr mtrix to find the highest corr features
   # Create correlation matrix
    # Retain upper triangular values of correlation matrix and
    # make Lower triangular values Null
    upper_corr_mat = corr_mat.where(
        np.triu(np.ones(corr_mat.shape), k=1).astype(np.bool))

    # Convert to 1-D series and drop Null values
    unique_corr_pairs = upper_corr_mat.unstack().dropna()

    # Sort correlation pairs
    sorted_mat = unique_corr_pairs.sort_values()
   
    print("sorted matrix")
    print(sorted_mat)
    corr=corr_mat 
    '''

       # Set up the matplotlib plot configuration
    #
    f, ax = plt.subplots(figsize=(12, 10))
    #
    # Generate a mask for upper traingle
    #
    mask = np.triu(np.ones_like(corr, dtype=bool))
    #
    # Configure a custom diverging colormap
    #
    cmap = sns.diverging_palette(230, 20, as_cmap=True)
    #
    # Draw the heatmap
    #
    sns.heatmap(corr, annot=True, mask = mask, cmap=cmap)
    plt.savefig('corrheatmap.png')
    plt.show()
    '''
    # min-max normalization
    if normalise == True:
        trans = MinMaxScaler()
        data_in = trans.fit_transform(data_in)
    if normalise == False:
        data_in=data_in.to_numpy()
    print('data_in')
    print(data_in)
        #data_inputx = data_in[:,0:8] # all features 0, - 7
        #data_inputx = data_in[:,[1]]  # one feature
    #data_inputx = data_in[:,[0,3]]  # two features   
    
    #data_in=data_in.to_numpy()
   
    #data_inputx = data_in[:, [1, 6]]#2 highly correlated features
    data_inputx = data_in[:,0:10] # all features 0-10
    print(data_inputx)
    #sns.pairplot(pd.DataFrame(data_inputx),hue='Rings', size=2.5)
    
    ######## create scatter plot
    
    plt.clf() 
    '''
    plt.rcParams["figure.figsize"] = [7.50, 3.50]
    plt.rcParams["figure.autolayout"] = True
    plt.scatter(data_inputx[:, 0], data_inputx[:, 2],color = 'hotpink')
    plt.plot(np.unique(data_inputx[:, 0]), np.poly1d(np.polyfit(data_inputx[:, 0], data_inputx[:, 2], 1))(np.unique(data_inputx[:, 0])))
    plt.xlabel('Shell Weight')
    plt.ylabel('Rings')
    plt.savefig('scatterplot1.png')
    plt.show()
    plt.clf() '''
    ########
      ######## create scatter plot 
    '''
    plt.clf() 
    plt.rcParams["figure.figsize"] = [7.50, 3.50]
    plt.rcParams["figure.autolayout"] = True
    plt.scatter(data_inputx[:, 1], data_inputx[:, 2], color = '#88c999')
    plt.plot(np.unique(data_inputx[:, 1]), np.poly1d(np.polyfit(data_inputx[:, 1], data_inputx[:, 2], 1))(np.unique(data_inputx[:, 1])))
    plt.xlabel('Diameter')
    plt.ylabel('Rings')
    plt.savefig('scatterplot2.png')
    plt.show()
    plt.clf() '''
    ########
    
    ######### plot histograms
    '''plt.bar(data_inputx[:, 0], data_inputx[:, 4])
    plt.xticks(data_inputx[:, 4])
    plt.yticks(data_inputx[:, 0]) #This may be included or excluded as per need
    plt.xlabel('Names')
    plt.ylabel('Probability')
    plt.show()'''
    '''
    x1 = data_inputx[:, 0]
    x2 = data_inputx[:, 1]
    x3 = data_inputx[:, 2]
    plt.clf()
    plt.hist(x1, bins=30, alpha=0.5, label='Shell Weight')
    plt.hist(x2, bins=30, alpha=0.5, label='Diameter')
    plt.hist(x3, bins=30, alpha=0.5, label='Rings')
    plt.legend(loc='upper right')
    plt.savefig('histogram.png')
    plt.show() '''
    #########
 

    ###############Create a 60/40 train/test split 
    data_inputy = data_in[:,7] # this is target - so that last col is selected from data
   # data_inputx = data_inputx[:,[0,1]] # the 2 independant variables
    
    #data_inputy = data_inputx[:,8] # this is target - so that last col is selected from data
    #data_inputx = data_inputx[:,[0,10]] # all features
    
    percent_test = 0.4
    '''testsize = int(percent_test * data_inputx.shape[0]) 
    x_train = data_inputx[:-testsize]
    x_test = data_inputx[-testsize:] 
    y_train = data_inputy[:-testsize]
    y_test = data_inputy[-testsize:]'''


      #another way you can use scikit-learn train test split with random state
    x_train, x_test, y_train, y_test = train_test_split(data_inputx, data_inputy, test_size=percent_test, random_state=i)

    return x_train, x_test, y_train, y_test
     
def scikit_linear_mod(x_train, x_test, y_train, y_test): 
 
    regr = linear_model.LinearRegression()

 
    # Train the model using the training sets
    regr.fit(x_train, y_train)

    # Make predictions using the testing set
    y_pred = regr.predict(x_test)
 
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))  
    rsquared = r2_score(y_test, y_pred) 
 
    residuals = y_pred - y_test
    plt.clf()
   # plt.plot(residuals, linewidth=1)
    plt.scatter(residuals,y_pred)
 
    plt.savefig('scikit_linear.png')
    plt.show()
    return rmse, rsquared, regr.coef_

 


def main(): 

    normalise = False
 
    
    max_exp = 30

    rmse_list = np.zeros(max_exp)
    rsq_list = np.zeros(max_exp)

    for i in range(0,max_exp):
        
        x_train, x_test, y_train, y_test = get_data(normalise, i)
        rmse, rsquared, coef = scikit_linear_mod(x_train, x_test, y_train, y_test)
        
        rmse_list[i] = rmse
        rsq_list[i] = rsquared 
        

    print(rmse_list)
    # Explained variance score: 1 is perfect prediction 
    print(rsq_list)
    
    mean_rmse = np.mean(rmse_list)
    std_rmse = np.std(rmse_list)

    mean_rsq = np.mean(rsq_list)
    std_rsq = np.std(rsq_list)

    print(mean_rmse, std_rmse, ' mean_rmse std_rmse')

    print(mean_rsq, std_rsq, ' mean_rsq std_rsq')
   

    plt.clf()
    plt.hist(rmse_list, bins=20, alpha=0.5, label='RMSE')
    plt.legend(loc='upper right')
    plt.savefig('histogram.png')
    plt.show()
    
  
    plt.clf()
    plt.hist(rsq_list, bins=20, alpha=0.5, label='RSQ')
    plt.legend(loc='upper right')
    plt.savefig('histogramrsq.png')
    plt.show()


if __name__ == '__main__':
     main()
```
